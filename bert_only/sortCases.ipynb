```python
import pickle

cases = pickle.load(open("processedCases.pickle", "rb"))
```

```python
cases[0]
```

```python
justice_counts = {}
for case in cases:
    for justice in case["votes"].keys():
        if justice not in justice_counts:
            justice_counts[justice] = 1
        justice_counts[justice] += 1
```

```python
import pandas as pd

df = pd.DataFrame(justice_counts.items(), columns=["justice", "count"]).sort_values("count")
df.plot(x="justice", y="count", kind="barh", figsize=(20,20))
```

```python
court_counts = {}
for case in cases:
    if case["court"] not in court_counts:
        court_counts[case["court"]] = 1
    court_counts[case["court"]] += 1
```

```python
courtDf = pd.DataFrame(court_counts.items(), columns=["court", "count"]).sort_values("count")
courtDf.plot(x="court", y="count", kind="barh", figsize=(20,20))
```
